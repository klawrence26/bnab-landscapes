```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from collections import defaultdict
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
    
from scipy.sparse import dok_matrix, csr_matrix
from sklearn.preprocessing import normalize
import networkx as nx
import itertools
from tqdm.notebook import tqdm
colorH1, colorH3, colorFluB, colorH9 = "#E8735C", "#72C2A6", "#5482A7", "#663399"

mpl.rcParams['figure.dpi'] = 300

#### antibody choice ####
antibody = "9114"
#antibody = "6261" 

antigens = ["H1", "H3", "FluB"] if antibody == "9114" else ["H1", "H9"]
L = (16 if antibody == "9114" else 11)
sequences_int = range(2**L)

if antibody == "9114":
    df = pd.read_csv("../../CR9114/Kd_meanbin/kd_processed/20210323_HA_unadj_fil_merg.csv",
                     dtype={"variant":"str"})
    df = df.rename(columns={"h1_mean": "H1_log10Kd", "h3_mean": "H3_log10Kd", "fluB_mean": "FluB_log10Kd",
                        "h1_sem": "H1_stelog10Kd", "h3_sem": "H3_stelog10Kd", "fluB_sem": "FluB_stelog10Kd"})
else:
    df = pd.read_csv("../../CR6261/Kd_meanbin/kd_processed/20210323_6261_HA_unadj_fil_merg.csv",
                     dtype={"variant":"str"})
    df = df.rename(columns={"h1_mean": "H1_log10Kd", "h9_mean": "H9_log10Kd",
                        "h1_sem": "H1_stelog10Kd", "h9_sem": "H9_stelog10Kd"})
df["variant_int"] = df.variant.apply(lambda x: int(x, 2))
df = df.set_index("variant_int").reindex(range(0, 2**L)).reset_index()
#for ag in antigens:
    #df[f"{ag}_log10Kd"] = -df[f"{ag}_log10Kd"]
    #df.loc[df[f"{ag}_log10Kd"].isna(), f"{ag}_log10Kd"] = -7 if (ag == "H1" or antibody == "6261") else -6
    #df.loc[df[f"{ag}_stelog10Kd"].isna(), f"{ag}_stelog10Kd"] = -0
for ii in range(16):
    df[f"mutation_{ii+1}"] = df.variant.str[ii] == '1'


df
    
```

```python


def mutation(s, ii, L):
    """ Return the sequence (in int format) with base s,
        with a somatic mutation at pos ii """
    return s | 2**(L-1-ii)

def nb_mutation(s):
    """ Number of mutations in variant s """
    return bin(s).count("1")

def reachable(s, L):
    return [(s | 2**ii) for ii in range(L) if (s | 2**ii) > s]

# "jump probability" function
def proba_markov(delta,N):
    if np.isnan(delta):
        return 0.0
    if delta == 0:
        return 1/N
    return (1 - np.exp(-delta))/(1 - np.exp(-N*delta))
    
    # if Δ <= 0:
    #     return 0# 1
    # else:
    #     return 1 - np.exp(-Δ)
```

```python
Nbootstrap = 1
N = 1000
# Create the transition matrix between variant, depending on context
df = df.sort_values(by=["variant_int"])
pmut, pmut_normed = {}, {}
for nb in tqdm(range(Nbootstrap)):
    for ag in antigens + ["all"]:
        pmut[ag, nb] = dok_matrix((2**L, 2**L), dtype=np.float64)
        if ag != "all":
            fitnesses = np.random.normal(
                        (df[f"{ag}_log10Kd"]).values.astype('float32'),
                        (df[f"{ag}_stelog10Kd"]).values.astype('float32'))
        else:
            fitnesses = sum([np.random.normal(
                        (df[f"{aa}_log10Kd"]).values.astype('float32'),
                        (df[f"{aa}_stelog10Kd"]).values.astype('float32')) for aa in antigens])/(len(antigens))
                         
        for s in sequences_int:
            tot = 0
            for t in reachable(s, L):
                delta = fitnesses[t] - fitnesses[s]
                ### formula transition probability ###
                pmut[ag, nb][s, t] = proba_markov(delta,N)
                    
        ### change format for something that can deal with multiplications
        pmut[ag, nb].tocsr()

        ### normalize the sparse matrix
        pmut_normed[ag, nb] = normalize(pmut[ag, nb], norm='l1', axis=1)
        
        
```

```python
if antibody == "9114":
    antigen_contexts = { "optimal": ["H1"]*5 + ["H3"]*8 + ["FluB"]* 3,
                         "H1 alone": ["H1"]*16,
                         "mixed": ["all"]*16 }
    antigen_context = antigen_contexts["mixed"]
elif antibody == "6261":
    antigen_contexts = { "H1": ["H1"]*11,
                         "H9": ["H9"]*11,
                         "mixed": ["all"]*11 }
    antigen_context = antigen_contexts["H9"]


## select random paths
nb = 10000
log_likelihoods = np.zeros(nb)
nb_zeros = np.zeros(nb)
if antibody == '6261':
    path_kds_H1 = np.zeros((nb,L+1))
    path_kds_H9 = np.zeros((nb,L+1))
elif antibody == '9114':
    path_kds_H1 = np.zeros((nb,L+1))
    path_kds_H3 = np.zeros((nb,L+1))
    path_kds_B = np.zeros((nb,L+1))

for ii in tqdm(range(nb)):
    mutations = np.random.choice(range(L), size=L, replace=False)
    var = 0
    path = [var]
    if antibody == '6261':
        path_kds_H1[ii,0] = df.iloc[var].H1_log10Kd
        path_kds_H9[ii,0] = df.iloc[var].H9_log10Kd
    elif antibody == '9114':
        path_kds_H1[ii,0] = df.iloc[var].H1_log10Kd
        path_kds_H3[ii,0] = df.iloc[var].H3_log10Kd
        path_kds_B[ii,0] = df.iloc[var].FluB_log10Kd
        

    for nmut, mut in enumerate(mutations):
        var = mutation(int(var), mut, L)
        if antibody == '6261':
            if np.isnan(df.iloc[var].H1_log10Kd):
                path_kds_H1[ii,nmut+1] = 7
            else:
                path_kds_H1[ii,nmut+1] = df.iloc[var].H1_log10Kd
            if np.isnan(df.iloc[var].H9_log10Kd):
                path_kds_H9[ii,nmut+1] = 7
            else:
                path_kds_H9[ii,nmut+1] = df.iloc[var].H9_log10Kd
        elif antibody == '9114':
            if np.isnan(df.iloc[var].H1_log10Kd):
                path_kds_H1[ii,nmut+1] = 7
            else:
                path_kds_H1[ii,nmut+1] = df.iloc[var].H1_log10Kd
            if np.isnan(df.iloc[var].H3_log10Kd):
                path_kds_H3[ii,nmut+1] = 6
            else:
                path_kds_H3[ii,nmut+1] = df.iloc[var].H3_log10Kd
            if np.isnan(df.iloc[var].FluB_log10Kd):
                path_kds_B[ii,nmut+1] = 6
            else:
                path_kds_B[ii,nmut+1] = df.iloc[var].FluB_log10Kd
            

        v = pmut_normed[antigen_context[nmut], 0][path[-1], var]
        if v == 0:
            nb_zeros[ii] += 1
        else:
            log_likelihoods[ii] += np.log(v)
        path += [var]

print(path_kds_H1)
```

```python
avg_path_H1 = np.average(path_kds_H1,weights = np.exp(log_likelihoods),axis=0)
avg_path_H3 = np.average(path_kds_H3,weights = np.exp(log_likelihoods),axis=0)
avg_path_B = np.average(path_kds_B,weights = np.exp(log_likelihoods),axis=0)

#avg_path_H1 = np.average(path_kds_H1,axis=0)

print(avg_path_H1.shape)
# sns.distplot(log_likelihoods[nb_zeros < 2])
# plt.xlabel("log-likelihood")
# plt.show()

plt.figure(figsize=(4,4))
plt.plot(range(L+1),avg_path_H1,color=colorH1)
plt.plot(range(L+1),avg_path_H3,color=colorH3)
plt.plot(range(L+1),avg_path_B,color=colorFluB)
plt.title('mixed')
plt.tight_layout()
plt.savefig('images/average_paths_9114_mixed.pdf')
plt.show()
```
