```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
sns.set_style('white')

import os
import collections
import glob
from scipy.stats import pearsonr
import scipy.optimize
import scipy.stats
```

```python
results = pd.DataFrame(pd.read_csv('../../../../CR9114/Isogenic_validation/Isogenic_Kd/Mutants/fitted_Kds.csv'))
results
```

```python
g = results.groupby(['antigen', 'shortname'])
results_merge = pd.DataFrame({
    'KD' : g.mean()['KD'],
    'KD_std' : g.std()['KD'],
    'KD_titeseq' : g['KD_titeseq'].mean(),
    'KD_titeseq_err' : g['KD_titeseq_err'].mean()
})

results_merge = results_merge.append(pd.DataFrame([[-6,np.nan,6,np.nan]], index=[('H3', 'ger')], columns=results_merge.columns))
results_merge = results_merge.append(pd.DataFrame([[-6,np.nan,6,np.nan]], index=[('FluB', 'ger')], columns=results_merge.columns))
results_merge
```

```python
sns.set_style('ticks')
from matplotlib.ticker import FixedLocator, FixedFormatter
#sns.set_context('notebook')
plot_list = results_merge[~results_merge['KD_titeseq'].isna()]
plt.rcParams['font.sans-serif'] = "Arial"
# Then, "ALWAYS use sans-serif fonts"
plt.rcParams['font.family'] = "sans-serif"
plt.rcParams['axes.linewidth'] = 0.5
flierprops = dict(marker='o', markersize=2)
plt.figure(figsize=(1.5,1.5))
x_locator = FixedLocator([6,7,8,9,10])
slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(-plot_list['KD'], plot_list['KD_titeseq'])
xs = np.array([6,10])
plt.plot(xs, xs, 'k-')
plt.plot(xs, slope*xs + intercept, 'k:')
corr, _ = pearsonr(-plot_list['KD'], plot_list['KD_titeseq'])

colordict = {'H1':'tomato', 'H3':'mediumaquamarine', 'FluB':'steelblue'}
# for i in plot_list.index:
#     plt.text(x=-plot_list['KD'][i] + 0.07,
#              y=plot_list['KD_titeseq'][i] - 0.05,
#              s=i[1],
#              va='center',
#              ha='left',
#              color=colordict[i[0]],
#              fontsize=10,
#              alpha=0.8
#             )
#plt.plot([6,10], [6,10], 'k:')
for ag,color in colordict.items():
    to_plot = plot_list.loc[ag,:]
    #plt.plot([], [], color, marker='o', ls='None', label=ag)
    plt.errorbar(x=-to_plot['KD'],
                 y=to_plot['KD_titeseq'],
                 xerr=to_plot['KD_std'],
                 yerr=to_plot['KD_titeseq_err'],
                 linestyle='None',
                 linewidth=1,
                 marker='o',
                 markersize=4,
                 markeredgecolor='k',
                 markeredgewidth=0.5,
                 label=ag,
                 ecolor='k',
                color=color)
import scipy.stats
plt.xticks(ticks=[6,7,8,9,10],labels=['6','7','8','9','10'],fontsize=7)
plt.yticks(ticks=[6,7,8,9,10],labels=['6','7','8','9','10'],fontsize=7)
#plt.legend(fontsize=7,bbox_to_anchor=(0.5, 1.1), loc='right')
plt.tick_params(length=2.5,width=0.5)
#plt.legend(loc='best',fontsize=7)
plt.xlabel(r'$-\log K_d$ isogenic FACS',fontsize=7,labelpad=0)
plt.ylabel(r'$-\log K_d$ Titeseq',fontsize=7,labelpad=0)
#plt.title('r =  %.2f' % corr,fontsize=8,y=1)
#plt.title('CR9114, Comparison of Titeseq to isogenic measurement\n$R^2=%0.3f$' % r_value**2)
plt.axis('square')
plt.savefig('9114validation.png',transparent=True, dpi=600,bbox_inches='tight')
print(corr)
```
