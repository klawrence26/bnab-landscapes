```python
mutants = ['germ','som','11516','4','46','456','6','71213','9','10','910','som3','som10']
means=[]
sems=[]
for mutant in mutants:
    mean = summary[summary['file'].str.contains("_%s_"%mutant)]['mean'].mean(axis=0)
    sem = summary[summary['file'].str.contains("_%s_"%mutant)]['mean'].sem(axis=0)
    means.append(mean)
    sems.append(sem)
lists2=zip(mutants,means,sems)
summary2 = pd.DataFrame(lists2, columns=['shortname','isogenic_exp_mean','isogenic_exp_sem'])
summary2.to_csv('isogenic_meanlogfluor_expression.csv')
summary2
```
